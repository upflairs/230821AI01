```python
vid = cv2.VideoCapture(0)
fd = cv2.CascadeClassifier(r'data/haarcascades/haarcascade_frontalface_default.xml')

pd = cv2.HOGDescriptor()
pd.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())

while True:
    ret, img = vid.read()
    if ret:
        igray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        fbox = fd.detectMultiScale(igray, 1.2, 4)
        
#         person, w = pd.detectMultiScale(img, winStride = (8,8))
#         for x,y,w,h in person:
#             cv2.rectangle(img, (x,y), (x+w, y+h), (0,0,255),5)
            
        if len(fbox):
            print(fbox)
            for x,y,w,h in fbox:
#                 img_face = img[y:y+h, x:x+w, :]
                cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,255), 10)
                cv2.putText(img, 'PEEYUSH SANAM', (x-20,y-20), fontFace = cv2.FONT_HERSHEY_COMPLEX, fontScale=1, color=(0,255,255))
#                 cv2.imwrite('image.png',img_face)
                cv2.imshow('preview',img)        
        else:
            cv2.imshow('preview',img)
#         cv2.imshow('preview',img)
    if cv2.waitKey(1) & 0xff == ord('s'):
        break
vid.release()
cv2.destroyAllWindows()
```
