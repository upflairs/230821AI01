```python
import cv2
import skimage as ski
```

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('dhoni.jpg')
```

```python
img_r = img[:,:,-1]
img_g = img[:,:,-2]
img_b = img[:,:,-3]
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from skimage import viewer
```

```python
view = viewer.ImageViewer(ski.io.imread('dhoni.jpg'))
view.show()
```

```python
# img_yellow = img_gray - img_b
img_yellow = cv2.subtract(img_gray,img_b)
```

```python
ski.io.imshow(img_yellow)
```

```python
view = viewer.ImageViewer(img_yellow)
view.show()
```

```python
_,img_yellow_bw = cv2.threshold(img_yellow,70,255,cv2.THRESH_BINARY)
plt.figure(figsize=(12,8))
ski.io.imshow(img_yellow_bw,cmap='gray')
```

# Morphology
1. Dilation
2. Erosion

3. Closing
4. Opening

```python
import numpy as np
```

```python
kern = np.ones((3,3), np.uint8)
```

```python
img_y_bw_1 = cv2.dilate(img_yellow_bw,kern, iterations=1)
# img_y_bw_1 = cv2.erode(img_y_bw_1,kern, iterations=1)
```

```python
plt.figure(figsize=(20,8))
plt.subplot(1,2,1)
ski.io.imshow(img_yellow_bw,cmap='gray')
plt.title('Original BW')

plt.subplot(1,2,2)
ski.io.imshow(img_y_bw_1,cmap='gray')
plt.title('Dilated BW')
```

```python
img_y_bw_2 = ski.morphology.remove_small_objects(img_y_bw_1.astype(bool),300,connectivity=1).astype(np.uint8)
```

```python
plt.figure(figsize=(28,8))
plt.subplot(1,3,1)
ski.io.imshow(img_yellow_bw,cmap='gray')
plt.title('Original BW')

plt.subplot(1,3,2)
ski.io.imshow(img_y_bw_1,cmap='gray')
plt.title('Dilated BW')

plt.subplot(1,3,3)
ski.io.imshow(img_y_bw_2,cmap='gray')
plt.title('No Noise BW')
```
