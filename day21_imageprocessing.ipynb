#### Image Color Space Conversion

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
img_gray.shape
```

```python
from skimage import color
```

```python
import skimage as ski
```

```python
img1_gray = ski.color.rgb2gray(img1)  # Y = 0.2125 R + 0.7154 G + 0.0721 B
img1_grey = ski.color.rgb2grey(img1)  # Y = 0.2125 R + 0.7154 G + 0.0721 B
```

```python
img1_gray.shape
```

```python
img1_grey.shape
```

```python
io.imshow(img1)
```

```python
plt.figure(figsize=(20,8))
plt.subplot(1,3,1)
io.imshow(img_gray)
plt.title('OpenCV Gray Version')
plt.subplot(1,3,2)
io.imshow(img1_gray)
plt.title('Skimage Gray Version')
plt.subplot(1,3,3)
io.imshow(img1_grey)
plt.title('Skimage Grey Version')
plt.show()
```

```python
th, img_bw = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY) # convert the pixels into black if the value is less than 100, else convert to white

io.imshow(img_bw)

img_bw.shape
```

```python
th, img_bin = cv2.threshold(img_gray, 150, 255, cv2.THRESH_BINARY) # convert the pixels into black if the value is less than 100, else convert to white

io.imshow(img_bin)

img_bin.shape
```

##### plane extraction

```python
#red plane
img_red_plane = img[:,:,-1]
img_green_plane = img[:,:,-2]
img_blue_plane = img[:,:,-3]
```

```python
plt.figure(figsize=(20,8))
plt.subplot(1,4,1)
io.imshow(img_red_plane)
plt.title('Red Plane')
plt.subplot(1,4,2)
io.imshow(img_green_plane)
plt.title('Green Plane')
plt.subplot(1,4,3)
io.imshow(img_blue_plane)
plt.title('Blue Plane')
plt.subplot(1,4,4)
io.imshow(img_gray)
plt.title('Gray Version')
plt.show()
```
